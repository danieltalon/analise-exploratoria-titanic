```python
# importa a biblioteca pandas
import pandas as pd 
import matplotlib.pyplot as plt 
import seaborn as sns

# url do arquivo de dados do Titanic
url = 'https://raw.githubusercontent.com/datasciencedojo/datasets/master/titanic.csv'

# carrega os dados em um DataFrame
df = pd.read_csv(url)

# exibe as 5 primeiras linhas para uma rapida visualização
df.head()
```

```python
df.info()
```

```python
df.describe()
```

```python
# configura estilo do gráfico
sns.set_theme(style="whitegrid")

# cria gráfico de contagem para a coluna 'survived'
# 0 = não sobreviveu, 1 = sobreviveu
plt.figure(figsize=(8,5)) # opcional: ajusta o tamanho da figura
sns.countplot(x='Survived',data=df)
plt.title('Contagem de sobreviventes vs. Não sobreviventes')
plt.show()
```

```python
# cria um gráfico de contagem para Pclass separado por Survived
palette_colors = {1:"skyblue", 0:"salmon"}
plt.figure(figsize=(10,6))
sns.countplot(x='Pclass',hue='Survived', data=df, palette=palette_colors)
plt.title('Contagem de sobreviventes por Classe Social', fontsize=16)
plt.legend(title='Sobreviveu', labels=['Não','Sim']) # valores em ordem crescente 0 (não) e 1 (sim)
plt.xlabel('Classe social (Pclass)', fontsize=12)
plt.ylabel('# Passageiros', fontsize=12)
plt.show()
```

```python
# cria um histograma para a coluna Age
# o argumento 'kde=true' adiciona uma linha de estimativa de densidade
plt.figure(figsize=(12,6))
# dropna remove valores nulos no gráfico
# bins = em quantos grupos a idade será separada  
sns.histplot(df['Age'].dropna(),bins=30,kde=True) 
plt.title('Distribuição de Idade dos Passageiros')
plt.xlabel('Idade')
plt.ylabel('Frequência')
plt.show()
```

```python
# Cria um gráfico de contagem para 'Sex', separado por 'Survived'
plt.figure(figsize=(10, 6))
sns.countplot(x='Sex', hue='Survived', data=df)
plt.title('Contagem de Sobrevivência por Sexo')
plt.legend(title='Sobreviveu', labels=['Não', 'Sim'])
plt.show()
```
